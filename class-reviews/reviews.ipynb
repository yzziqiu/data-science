```python
text0=data.iloc[:,0] #提取所有数据
text1=[i.astype('U13') for i in text0]
```

```python
senti=[SnowNLP(i).sentiments for i in text1] 
newsenti=[]
for i in senti:
  if (i>=0.6):
      newsenti.append(1)
  else:
      newsenti.append(-1)
data['predict']=newsenti  #将新的预测标签增加为text的某一列，所以现在text的第0列为评论文本，第1列为实际标签，第2列为预测标签
counts=0
for j in range(len(data.iloc[:,0])): #遍历所有标签，将预测标签和实际标签进行比较，相同则判断正确。
    if data.iloc[j,2]==data.iloc[j,1]:
        counts+=1
print(float(counts)/float(len(data)))#输出本次预测的准确率
```
