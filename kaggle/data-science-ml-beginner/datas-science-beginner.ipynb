## Data Science Toolbox

```python
# tuple
def tuble_ex():
    """ return defined t tuble"""
    t = (1,2,3)
    return t
a,b,c = tuble_ex()
print(a,b,c)
```

```python
# scope
x = 2
def f():
    x = 3
    return x
print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
# First local scopesearched, then global scope searched
# if two of them cannot be found lastly built in scope searched.
```

```python
# How can we learn what is built in scope
import builtins
dir(builtins)
```

```python
# nested functions
# There is a LEGB rule that is search local scope
# enclosing function, global and built in scopes, respectively
#nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
```

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():               # If you do not understand this part turn for loop part and look at dictionary in for loop
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

```python
# lambda function
square = lambda x: x**2     # where x is name of argument
print(square(4))
tot = lambda x,y,z: x+y+z   # where x,y,z are names of arguments
print(tot(1,2,3))
```

```python
# anonymous function
# map(func,seq) : applies a function to all the items in a list
number_list = [1,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
# iterators
# iteration example
name = "ronaldo"
it = iter(name)
print(next(it))    # print next iteration
print(*it)         # print remaining iteration
```

```python
# zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip) # unzip returns tuble
print(un_list1)
print(un_list2)
print(type(un_list2))
```

```python
# list
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```

```python
df.pivot(index="treatment",columns = "gender",values="response")
```

```python
#STACKING and UNSTACKING DATAFRAME
# deal with multi label indexes
# level: position of unstacked index
# swaplevel: change inner and outer level index position
df1 = df.set_index(["treatment","gender"])
df1
# lets unstack it
```

```python
# level determines indexes
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
# change inner and outer level index position
df2 = df1.swaplevel(0,1)
df2
```
